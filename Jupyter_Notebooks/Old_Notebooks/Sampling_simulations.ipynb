```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 8000
sample_sizes = [100, 250, 400]
n_I = 1
k = 200
T = 11
n_I = np.loadtxt('../Text_files/Stochastic/Networks/barabasi-albert/deterministic_R03.0.txt')[1,:].astype('int')
print(n_I)
fig, ax = plt.subplots(figsize=(10,6))
for m in sample_sizes:
    probs_dect = np.array([])
    probs_no_dect = np.array([])
    probs = np.array([])
    ## run for T days
    for t in range(T):
        counter = 0
        ## Run k samples
        for i in range(k):
            sample = np.array(random.sample(range(N-min(t,7)*m),m))
            ## if detected
            if (sum(np.isin(sample,range(n_I[t])))):
                counter +=1
        probs_dect = np.append(probs_dect,(counter/k))
        probs_no_dect = np.append(probs_no_dect, 1-(counter/k))
        probs = np.append(probs, probs_dect[t]*np.prod(probs_no_dect[:-1]))
    ax.plot(range(T), np.cumsum(probs), '.-', label = 'm=%d'%m)
ax.hlines(1, 0, T, label = '90%', color = 'k', linestyle = 'dashed')
ax.hlines(0.9, 0, T, label = '90%', color = 'brown', linestyle = 'dashed')
ax.legend(fontsize = 16)
ax.set_xlabel('time [days]', fontsize = 16)
ax.set_ylabel('Cum. Prob. of detecting at least one', fontsize = 16)
ax.set_xlim(0,T-1)
```
